```python
# Demo Tuition Fee Management System (Console Version)

months = ["January","February","March","April","May","June",
          "July","August","September","October","November","December"]

students = {}

def add_student():
    name = input("Enter student name: ")
    fee_input = input("Enter monthly fee: ")
    # Remove 'Rs ' and any spaces, then convert to float
    fee = float(fee_input.replace('Rs ', '').strip())
    students[name] = {"fee": fee, "payments": {m: False for m in months}}
    print(f"Student {name} added successfully.\n")

def record_payment():
    name = input("Enter student name: ")
    if name not in students:
        print("Student not found.\n")
        return
    print("Months:", ", ".join(months))
    month = input("Enter month for payment: ")
    if month not in months:
        print("Invalid month.\n")
        return
    if students[name]["payments"][month]:
        print("Payment already recorded.\n")
    else:
        students[name]["payments"][month] = True
        print(f"Payment for {month} recorded for {name}.\n")

def view_record():
    name = input("Enter student name: ")
    if name not in students:
        print("Student not found.\n")
        return
    print(f"\nPayment record for {name}")
    print("-------------------------")
    total_paid = 0
    for m, paid in students[name]["payments"].items():
        status = "Paid" if paid else "Pending"
        print(f"{m}: {status}")
        if paid: total_paid += 1
    total_due = (12 - total_paid) * students[name]["fee"]
    print(f"\nMonths Paid: {total_paid}/12")
    print(f"Total Due: {total_due}\n")

def list_students():
    if not students:
        print("No students found.\n")
        return
    print("Students List:")
    for s in students:
        print(s)
    print()

while True:
    print("Tuition Fee Management System")
    print("1. Add Student")
    print("2. Record Payment")
    print("3. View Student Record")
    print("4. List Students")
    print("5. Exit")
    choice = input("Enter choice: ")
    if choice == "1":
        add_student()
    elif choice == "2":
        record_payment()
    elif choice == "3":
        view_record()
    elif choice == "4":
        list_students()
    elif choice == "5":
        print("Exiting...")
        break
    else:
        print("Invalid choice.\n")
```
